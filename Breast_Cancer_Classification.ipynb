**Breast Cancer Classification with a simple Neural Network**

# 1. Importing Required Libraries

```python
# Importing the required libraries for data manipulation, visualization, machine learning, and model evaluation

import numpy as np                  # Numerical computations
import pandas as pd                 # Data manipulation and analysis
import matplotlib.pyplot as plt     # Data visualization
import sklearn.datasets             # Load the Breast Cancer dataset
from sklearn.model_selection import train_test_split   # Split dataset into training and testing sets
from sklearn.metrics import accuracy_score             # Calculate model accuracy
```

# 2. Loading the Breast Cancer Dataset


```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# Display the complete Breast Cancer dataset
print(breast_cancer_dataset)
```

# 3.  Data Collection & Preprocessing

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

# 4.  Exploratory Data Analysis (EDA)

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

# 5.  Checking Missing Values

```python
# checking for missing values
data_frame.isnull().sum()
```

# 6.  Statistical Summary of the Dataset

```python
# statistical measures about the data
data_frame.describe()
```

# 7. Understanding the Target Variable

```python
# checking the distribution of Target Varibale
data_frame['label'].value_counts()
```

1 --> Benign

```python
# Group the dataset by tumor class (label) and compute the average of each feature
data_frame.groupby('label').mean()
```

# 8. Feature and Target Separation

```python
# Extract the input features by removing the target column
X = data_frame.drop(columns='label', axis=1)

# Extract the target labels (Benign = 1, Malignant = 0)
Y = data_frame['label']
```

```python
print(X)
```

```python
# Display the target variable (0 = Malignant, 1 = Benign)
print(Y)
```

# 9. Splitting Data into Training and Testing Sets

```python
# Split the input features and target labels into training (80%) and testing (20%) datasets
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=2
)
```

```python
# Display the number of samples and features in the original, training, and testing datasets
print(X.shape, X_train.shape, X_test.shape)
```

# 10. Feature Scaling using StandardScaler

```python
# Import StandardScaler to standardize the feature values before training the model
from sklearn.preprocessing import StandardScaler
```

```python
# Standardize the training and testing data using StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform it
X_train_std = scaler.fit_transform(X_train)

# Transform the testing data using the same scaler
X_test_std = scaler.transform(X_test)
```

**Building the Neural Network**

download.png

```python
# Import TensorFlow and Keras libraries for building and training the Artificial Neural Network (ANN)
import tensorflow as tf

# Set a random seed for reproducible results
tf.random.set_seed(3)

# Import the Keras API from TensorFlow
from tensorflow import keras
```

# 11. Building the Artificial Neural Network (ANN)

```python
# Build the Artificial Neural Network (ANN) by defining its layers

model = keras.Sequential([
    keras.Input(shape=(30,)),                     # Input layer
    keras.layers.Flatten(),
    keras.layers.Dense(20, activation='relu'),   # Hidden layer
    keras.layers.Dense(2, activation='sigmoid')  # Output layer
])
```

# 12. Compiling the Neural Network

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

# 13. Training the Model

```python
# training the Meural Network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

# 14. Visualizing Training & Validation Accuracy

```python
# Plot the training and validation accuracy over each training epoch
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

# Add title and axis labels to the plot
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')

# Display the legend to distinguish training and validation accuracy
plt.legend(['Training Data', 'Validation Data'], loc='lower right')

# Display grid lines and show the plot
plt.grid(True)
plt.show()
```

# 15. Visualizing Training & Validation Loss

```python
# Plot the training and validation loss over each training epoch
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

# Add title and axis labels to the plot
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')

# Display the legend to distinguish training and validation loss
plt.legend(['Training Data', 'Validation Data'], loc='upper right')

# Display grid lines and show the plot
plt.grid(True)
plt.show()
```

# 16. Evaluating Model Performance on Test Data

```python
# Evaluate the model's performance by calculating the test loss and accuracy
loss, accuracy = model.evaluate(X_test_std, Y_test)

# Print the accuracy achieved on the testing dataset
print(accuracy)
```

```python
# Display the dimensions of the standardized testing dataset
print(X_test_std.shape)

# Display the standardized feature vector of the first sample in the testing dataset
print(X_test_std[0])
```

# 17. Making Predictions on Test Samples

```python
# Use the trained ANN model to predict the class probabilities for the testing dataset
Y_pred = model.predict(X_test_std)
```

```python
# Display the dimensions of the prediction array
print(Y_pred.shape)

# Display the predicted class probabilities for the first sample in the testing dataset
print(Y_pred[0])
```

model.predict() gives the prediction probability of each class for that data point

```python
# Demonstrate the use of np.argmax() to identify the index of the highest value in a list
my_list = [0.25, 0.56]

# Compute the index of the maximum value
index_of_max_value = np.argmax(my_list)

# Display the input list
print(my_list)

# Display the index corresponding to the maximum value
print(index_of_max_value)
```

# 18. Converting Prediction Probabilities into Class Labels

```python
# Convert the predicted probability scores into their corresponding class labels
# np.argmax() returns the index of the highest probability for each prediction
Y_pred_labels = [np.argmax(i) for i in Y_pred]

# Display the predicted class labels for the testing dataset
print(Y_pred_labels)
```

# 20. Predicting Whether the Tumor is Benign or Malignant

```python
# Define the input data containing the feature values of a tumor sample
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,
              0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,
              0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,
              0.2433,0.06563)

# Convert the input data into a NumPy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the array to represent a single sample with multiple features
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Convert the reshaped array into a DataFrame using the original feature names
input_data_df = pd.DataFrame(input_data_reshaped, columns=X.columns)

# Standardize the input data using the previously fitted StandardScaler
input_data_std = scaler.transform(input_data_df)

# Predict the class probabilities for the input sample
prediction = model.predict(input_data_std)

# Display the predicted class probabilities
print(prediction)

# Convert the predicted probabilities into the corresponding class label
prediction_label = [np.argmax(prediction)]

# Display the predicted class label
print(prediction_label)

# Display the prediction result in a user-friendly format
if prediction_label[0] == 0:
    print('The tumor is Malignant')
else:
    print('The tumor is Benign')
```
